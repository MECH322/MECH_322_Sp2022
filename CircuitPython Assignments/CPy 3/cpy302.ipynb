# CPy 3.02 - NeoPixel Temperature Display

## Instructions

Use commands from the **`adafruit_circuitplayground`** module to create a temperature display on the CPB board that uses NeoPixels to display the temperature as sensed from the built-in temperature sensor. Use the built-in push buttons and slide switch to change the temperature display.

Continuously read the temperature from the built-in temperature sensor. Just before your main loop and then every 5 seconds (without using **`time.sleep()`**) display the temperature in Celsius and Fahrenheit with units (both rounded to one decimal place) on the same line of the REPL.

Every 2.5 seconds (without using **`time.sleep()`**) light up the number of NeoPixels in blue that correspond to the temperature's "tens" place. A tens place value of 0 should light up one (the zeroth index) NeoPixel. (See the numbering in the included image below.) A tens place value of 7 should light up the first 8 NeoPixels. For the next 2.5 seconds (again, no **`time.sleep()`**) light up the number of NeoPixels corresponding to the "ones" place (rounded to the whole number) in green. A ones value of 0 should light up one (the zeroth index) NeoPixel. A ones value of 9 should light up all 10 NeoPixels. When the slide switch is in the left position use Fahrenheit for the NeoPixel display and turn on the built-in red LED. When the slide switch is in the right position use Celsius and turn off the built-in red LED.

Make it so the built-in buttons temporarily override the cycling NeoPixel display. If just button A is pressed and held display the "tens" value. If just button B is pressed and held, display the "ones" value. When neither is pressed, the cycling between "tens" and "ones" should continue.

As an example, if the temperature is 73.2 Fahrenheit and the slide switch is in the left position, then the built-in red LED will be lit. Every 2.5 seconds the NeoPixels will cycle between 8 NeoPixels being illuminated blue and 4 NeoPixels being illuminated green.

Use a low brightness, like 0.05 for the NeoPixels.
